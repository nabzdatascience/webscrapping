```python
import pandas as pd
import requests as requests
from bs4 import BeautifulSoup
```

```python
url="https://www.flipkart.com/search?q=mobile%20phones%20under%2020000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
requested_page=requests.get(url)
```

```python
print(requested_page.text)
```

```python
soup=BeautifulSoup(requested_page.text,"html.parser")
```

```python
print(soup)
```

```python
titles=[]

```

```python
for mobiles in soup.find_all("div",class_="_4rR01T"):
    titles.append(mobiles.text)
```

```python
titles
```

```python
raw_data={'Mobile_names':titles}
```

```python
df=pd.Series(raw_data)
df
```

```python
df.to_csv('data.csv')
print(df)
```
